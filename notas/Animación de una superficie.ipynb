```julia
using PyPlot
using PyCall

@pyimport matplotlib.animation as anim
@pyimport numpy as np
```

```julia
fig = plt.figure()
ax = fig[:add_subplot](111, projection="3d", xlim=(-1,1), ylim=(-1,1), zlim=(0,1))

x = -2:0.1:2
y = -2:0.1:2

X,Y = np.meshgrid(x,y)
psurf = ax[:plot_surface](X, Y, np.zeros_like(X))

f(α) = exp(-α * (x.^2 .+ y'.^2))

α = [0.2:0.1:3]

function animate(i)
    ax[:cla]() # Clear axis = clear frame
    psurf = ax[:plot_surface](X, Y, f(α[i+1]), rstride=1, cstride=1) # mesh size depends on rstride and cstride
    ax[:grid](false) # temporary "fix" for this bug https://github.com/matplotlib/matplotlib/issues/750/
    ax[:set_zlim](0,1) # Keep it fixed so that it does not change in every frame
    return (psurf, None)
end

myanim = anim.FuncAnimation(fig, animate, frames=29, interval=10)
myanim[:save]("surface.gif", writer="imagemagick", fps=5)
```
