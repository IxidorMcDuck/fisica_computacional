```julia
using Images
using FixedPointNumbers
```

```julia
img = imread("region.pbm")
#show(img)
```

```julia
domain = int(data(reinterpret(Ufixed8, img))');
nx, ny = size(domain)
```

```julia
# indices of the interior points, i.e. where domain == 1 (white)
int_i, int_j = findn(domain)
```

```julia
# two's complement of the domain
c_domain = (domain+ones(size(domain)))%2

# indices of the boundary points, i.e. where domain == 0 (black)
b_i, b_j = findn(c_domain)
```

```julia
function f(x,y)
    if x == 0.0 || y == 0.0
        return 2.0
    elseif x == 1.0 || y == 1.0
        return 1.0
    end
end

#f(x,y) = 1.0

function Dirichlet(i, j, nx, ny)
    x = linspace(0,1,nx)
    y = linspace(0,1,ny)
    return f(x[i], y[j])
end
```

```julia
u = zeros(size(domain));
```

```julia
# For the boundary points we already know the solution
for n in 1:length(b_i)
    u[b_i[n],b_j[n]] = Dirichlet(b_i[n], b_j[n], nx, ny)
end
```

```julia
using PyPlot
```

```julia
matshow(u, cmap=get_cmap("Blues"), extent=(0,16,16,0))
xticks([0:16]);
yticks([0:16]);
grid()
```

```julia
function Laplace1D(N)
    # N: number of interior points
    # For a function of one variable the diag has 2's
    # For a function of two variables the diag has 4's
    full(SymTridiagonal(4*ones(N), -1*ones(N-1)))
end
```

```julia
function Laplace2D(N)
    # N: number of points in the horizontal direction
    kron(eye(N),Laplace1D(N)) + kron(full(SymTridiagonal(zeros(N), -1*ones(N-1))), eye(N))
end
```

```julia
A = Laplace2D(14);
```

```julia
b = zeros(14,14);
b[1,:] = [u[1,i] for i in 2:15]
b[1,1] += u[2,1]
b[1,14] += u[2,16]

for i in 2:13
    b[i,1] = u[i,1]
    b[i,14] = u[i,16]
end

b[14,:] = [u[16,i] for i in 2:15]
b[14,1] += u[15,1]
b[14,14] += u[15,16]
b = reshape(b, (14*14,1));
```

```julia
x = A\b

for n in 1:length(int_i)
    u[int_i[n],int_j[n]] = x[n]
end
```
