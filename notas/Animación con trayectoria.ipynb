```julia
using PyCall
using PyPlot

@pyimport matplotlib.animation as anim 
```

```julia
function euler(y0,t0,h,f,N)
    y = Vector{Float64}[[y0]]
    t = Float64[t0]
    for i in 1:N
        y_nueva = y0 + h*f(t0, y0)
        t0 = t0 + h
        y0 = y_nueva
        push!(t,t0)
        push!(y,[y0])
    end
    t,hcat(y...)'
end

function oscilador(t,y)
    θ = y[1]
    ω = y[2]
    [ω, -θ]
end

θ_0 = 1.0
ω_0 = 0.0
y_0 = [θ_0,ω_0]
t_0 = 0.0
h = 0.1
N = 100
f = oscilador

t,y = euler(y_0, t_0, h, f, N);
θ = y[:,1];
ω = y[:,2];
```

```julia
# Configuración del lienzo
fig = figure()
ax = plt.axes(xlim=(-2, 2), ylim=(-2, 2))

global particle = ax[:plot]([], [], "bo", lw=2)[1]
global trail = ax[:plot]([], [], "b-", lw=1)[1]

# Inicialización
function init()
    global particle
    global trail

    particle[:set_data]([], [])
    trail[:set_data]([], [])

    return ([particle, trail],None)
end

# Esta función dibuja lo que aparece en cada cuadro de la animación
# i es el número de cuadro o el n-ésimo paso de tiempo
# Por ejemplo, puede ser el número de paso en el método de Euler
# La función es llamada secuencialmente por anim.FuncAnimation para cada cuadro
function animate(i)
    global particle
    global trail

    # i+1 para que i=0 sea el índice uno
    trail[:set_data](θ[1:i+1], ω[1:i+1])
    particle[:set_data](θ[i+1], ω[i+1])
    return ([particle, trail],None)
end
```

```julia
# Generamos la animación.  blit=True significa que sólo redibuja lo que cambió de un cuadro a otro
myanim = anim.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=true);

# Guardamos la animación en un archivo .mp4. Listo para subir a YouTube
myanim[:save]("sinplot.mp4", extra_args=["-vcodec", "libx264"])

# Mostramos la animación aquí en el cuadernobb
display("text/html", string("""<video autoplay controls><source src="data:video/mp4;base64,""",
                            base64(open(readbytes,"sinplot.mp4")),"""" type="video/mp4"></video>"""))
```

Para que se vea el video, en Ubuntu puede ser necesario instalar `ubuntu-restricted-extras` y en Firefox configurar `media.gstreamer.enabled` como **true** en about:config.

También podemos guardar la animación en formato GIF e incluirla fácilmente usando Markdown.

```julia
myanim = anim.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=true);
myanim[:save]("trayectoria.gif", writer="imagemagick", fps=30);
```
