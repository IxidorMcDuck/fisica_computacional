```julia
using PyCall
using PyPlot

@pyimport matplotlib.animation as anim
@pyimport numpy.random as nprand
```

```julia
# Configuración del lienzo
fig = figure()
ax = plt.axes(xlim=(0, 2), ylim=(0, 2))

global img = ax[:imshow](nprand.rand(16,16), cmap=get_cmap("autumn"))

# Esta función dibuja lo que aparece en cada cuadro de la animación
# i es el número de cuadro o el n-ésimo paso de tiempo
# Por ejemplo, puede ser el número de paso en el método de Euler
# La función es llamada secuencialmente por anim.FuncAnimation para cada cuadro
function animate(i)
    global img
    # Aquí deben poner una matriz que resulta de resolver, por ejemplo, la ecuación del calor
    img[:set_array](nprand.rand(16,16))
    return (img, None)
end
```

```julia
# Generamos la animación.  blit=True significa que sólo redibuja lo que cambió de un cuadro a otro
myanim = anim.FuncAnimation(fig, animate, frames=100, interval=20, blit=true);

# Guardamos la animación en un archivo .mp4. Listo para subir a YouTube
myanim[:save]("density_plot.mp4", extra_args=["-vcodec", "libx264"])

# Mostramos la animación aquí en el cuaderno
display("text/html", string("""<video autoplay controls><source src="data:video/mp4;base64,""",
                            base64(open(readbytes,"density_plot.mp4")),"""" type="video/mp4"></video>"""))
```

Para que se vea el video, en Ubuntu puede ser necesario instalar `ubuntu-restricted-extras` y en Firefox configurar `media.gstreamer.enabled` como **true** en about:config.

También podemos guardar la animación en formato GIF e incluirla fácilmente usando Markdown.

```julia
myanim = anim.FuncAnimation(fig, animate, frames=100, interval=20, blit=true);
myanim[:save]("density_plot.gif", writer="imagemagick", fps=30);
```
