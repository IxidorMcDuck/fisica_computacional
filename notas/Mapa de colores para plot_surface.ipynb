PyPlot trae predefinidos estos [mapas de colores](http://matplotlib.org/examples/color/colormaps_reference.html).

Algunos compañeros me han preguntado cómo crear su propio mapa de colores para usar con `plot_surface`.

[Aquí](http://colormap.org/) pueden crear su mapa de colores interactivamente.

1. Copien el mapa de colores y lo pegan en un editor de texto, por ejemplo Gedit, como `paleta = [1,2,3;4,5,6;...]`.

2. Usen la función de buscar y reemplazar para cambiar las comas por espacios.

3. Guarden el archivo con el nombre `paleta.jl` en el mismo lugar donde lo vayan a usar en IJulia.

```julia
using PyPlot
using PyCall
@pyimport matplotlib.colors as mpc
```

```julia
include("paleta.jl")
azul_mar = mpc.ListedColormap(paleta/255)
```

```julia
register_cmap("azul_mar", azul_mar)
```

```julia
# Superficie en coordenadas cilíndricas

n = 50      # Discretización del eje z

theta = pi * linspace(0, 1, n)
rho = sin(theta)
zz = cos(theta)
phi = 2pi * linspace(0, 1, n)'

x = rho * cos(phi)
y = rho * sin(phi)
z = zz[:, ones(n)]   #  Tony's trick!   Who is Tony?  No idea. Jajaja

mapa = get_cmap("azul_mar")
plot_surface(x, y, z, rstride=1, cstride=1, cmap=mapa,
linewidth=0, antialiased=true, alpha=1);
```

### Colorear la esfera en términos de una función escalar

```julia
# Superficie en coordenadas cilíndricas

n = 50      # Discretización del eje z

theta = pi * linspace(0, 1, n)
rho = sin(theta)
zz = cos(theta)
phi = 2pi * linspace(0, 1, n)'

x = rho * cos(phi)
y = rho * sin(phi)
z = zz[:, ones(n)]   #  Tony's trick!   Who is Tony?  No idea. Jajaja

paleta = ("y", "b") # [y]ellow, [b]lue
xlen = n
ylen = n
colores = fill("",xlen,ylen)
for i in range(1,xlen)
    for j in range(1,ylen)
        colores[i,j] = paleta[(i+j)%length(paleta)+1]
    end
end

plot_surface(x, y, z, rstride=1, cstride=1, facecolors=colores,
linewidth=0, antialiased=true, alpha=1);
```
