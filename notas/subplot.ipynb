```julia
using PyPlot
```

```julia
t = [0:0.01:10]

subplots_adjust(hspace=0.0) # separación entre las gráficas

subplot(311)
ax1 = gca()
setp(ax1[:get_xticklabels](),visible=false) # No mostrar valores del eje x
ylabel(L"$\sin(t)$")
plot(t,sin(t),"r-")
axis("equal")

subplot(312, sharex=ax1)
ax2 = gca()
setp(ax2[:get_xticklabels](),visible=false) # No mostrar valores del eje x
ylabel(L"$\cos(t)$")
xlabel(L"$t$")
plot(t,cos(t))
axis("equal")

subplot(313, sharex=ax2)
ylabel(L"$j_0(t)$")
xlabel(L"$t$")
plot(t,besselj0(t),"g-")
axis("equal")
```
