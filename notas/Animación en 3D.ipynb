```julia
using PyPlot
using PyCall

@pyimport matplotlib.animation as anim 
```

```julia
fig = plt.figure()
ax = subplot(111, projection="3d", xlim=(-1,1), ylim=(-1,1), zlim=(0,2*pi))

global line = ax[:plot]([], [], [], "bo", lw=2)[1]

function init()
    global line

    line[:set_data]([], [])
    line[:set_3d_properties]([])

    return (line,None)
end

function animate(i)
    global line

    x = cos(2*pi*i/100)
    y = sin(2*pi*i/100)
    z = i*2*pi/100
    
    line[:set_data](x,y)
    line[:set_3d_properties](z)
    
    return (line,None)
end
```

```julia
myanim = anim.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=true);
myanim[:save]("3d.gif", writer="imagemagick", fps=30);
```
