# Método de Euler

$$
y_{n+1} = y_{n} + hf(t_{n},y_{n})
$$

- Es **explícito**: $y_{n+1}$ depende de un valor conocido, $y_{n}$
- Aproxima la derivada con una diferencia finita **adelantada**: 
$$
f(t_n, y_n) = \dot{y}(t_{n}) \approx \frac{y_{n+1} - y_{n}}{h}
$$
- $\tau_n := \frac{y_{n+1} - y_{n}}{h} - \dot{y}(t_{n})  = \frac{h}{2}\ddot{y}(\xi_n)$, $\xi_n \in (t_{n},t_{n+1})$ es el **error local de truncamiento**.
- Es un método de **primer orden**: $\tau_{n} = \mathcal{O}(h)$

### Para resolver una sola ecuación

Usamos Julia porque podemos escribir un programa **muy parecido** a las expresiones matemáticas que escribimos a mano. Es decir, los programas son fáciles de entender y se escriben en poco tiempo. En otras palabras, no se compliquen!

¿Qué argumentos debe llevar la función? Los ingredientes necesarios para usar el método de Euler:

- tiempo inicial: $t_0$
- tamaño del paso: $h$
- condición inicial: $y_0$
- función que determina la ecuación diferencial: $f$
- número de pasos que damos con el método de Euler: $N$

```julia
function euler(y0,t0,h,f,N)
    y = Float64[y0]
    t = Float64[t0]
    for i in 1:N
        y_nueva = y0 + h*f(t0, y0)
        t0 = t0 + h
        y0 = y_nueva
        push!(t,t0)
        push!(y,y0)
    end
    t,y
end
```

### Para resolver un sistema de ecuaciones es casi la misma función

Para el sistema de $m$ ecuaciones
$$
\dot{\vec{y}} = \vec{f}(t,\vec{y})
$$

El método de Euler está dado por

$$
\begin{array}{rcl}
y_{1,n+1} &=& y_{1,n} + hf_1(t_n,\vec{y}_{n})\\
y_{2,n+1} &=& y_{2,n} + hf_2(t_n,\vec{y}_{n})\\
&\vdots&\\
y_{m,n+1} &=& y_{m,n} + hf_m(t_n,\vec{y}_{n})\\
\end{array}
$$

```julia
function euler(y0,t0,h,f,N)
    y = Vector{Float64}[[y0]]
    t = Float64[t0]
    for i in 1:N
        y_nueva = y0 + h*f(t0, y0)
        t0 = t0 + h
        y0 = y_nueva
        push!(t,t0)
        push!(y,[y0])
    end
    t,hcat(y...)'
end
```

$y$ es un vector de vectores: $[[2.3, 3.1],[0.1,-4],...]$. Usamos `hcat` para convertirlo en una matriz. `hcat` con**cat**ena los vectores **h**orizontalmente:
$$
\mat{2.3 & 0.1 \\ 3.1 & -4}
$$
Cada renglón de esta matriz corresponde con una componente de $\vec{y}$. $\red{OJO}$: los renglones en Julia son tipo `Array{Float64,2}`. Ese tipo de dato no es conveniente para graficar con PyPlot. Por eso usamos la transpuesta `'`. Para que las columnas de la matriz sean las componentes de $\vec{y}$. Las columnas en Julia son vectores, es decir, tipo `Array{Float64,1}`.

### Oscilador armónico

```julia
function oscilador(t,y)
    θ = y[1]
    ω = y[2]
    [ω, -θ]
end
```

Para definir $\vec{f}(t,\vec{y})$ tenemos que hacer una convención sobre el significado de las entradas del vector. Aquí usamos la convención de que la primera entrada corresponde a $\theta$ y la segunda entrada corresponde a $\omega$. Por lo tanto,
$$
\vec{f}(t,\vec{y}) = \mat{\dot{\theta} \\ \dot{\omega}} = \mat{\omega \\ -\theta}
$$

Para dar la condición inicial $\vec{y}_0$ tenemos que usar la misma convención:
$$
\vec{y}_0 = \mat{\theta_{0} \\ \omega_{0}}
$$

```julia
θ_0 = 1.0
ω_0 = 0.0
y_0 = [θ_0,ω_0]
t_0 = 0.0
h = 0.01
N = 10000
f = oscilador

t,y = euler(y_0, t_0, h, f, N);
θ = y[:,1];
ω = y[:,2];
```

```julia
energia_Euler = 1/2*(θ.^2 + ω.^2);
energia_exacta = 1/2*(θ_0^2 + ω_0^2)*ones(length(energia_Euler));
```

```julia
using PyPlot
```

```julia
plot(t,energia_Euler - energia_exacta);
```

### Error global

$e_n = |y(t_{n}) - y_n|$, donde $y_n$ es la solución aproximada obtenida con el método de Euler en el paso $n$.

$$
\begin{array}{rrl}
&y_{n+1} &=& y_{n} + hf(t_{n},y_{n})\\
-&&\\
&y(t_n+h) &=& y(t_n) + hf(t_n, y(t_n)) + \frac12 h^2 \ddot{f}(\xi_n)\\
&&=&\frac12 h^2 \ddot{f}(\xi_n)
\end{array}
$$

Entonces, el error en un paso de Euler es $e(h) = \mathcal{O}(h^2)$.

El error acumulado desde $t=0$ hasta $t=nh=T$ es 
$$
e_n = ne(h) = \frac{T}{h}\frac{h^2}{2}\ddot{f}(\xi_n) = \mathcal{O}(h).
$$

```julia
θ_exacta = cos(t);
error_global = abs(θ - θ_exacta);
```

```julia
plot(t,error_global);
```

El error acumulado en el tiempo $t = 100$ NO es $\mathcal{O}(0.01)$!! ¿Por qué? Porque el método no es estable. El jueves discutiremos la estabilidad.
