# Euler

Método de Euler:

"Avanza en pasitos; desde donde estoy, sigue una recta con pendiente igual a la derivada evaluada ahí"

Empezamos en $x(t_0)=x_0$ en el tiempo $t_0$

Idea del método:

Dar un "paso" de tamaño $h$ en el tiempo: queremos saber cuánto vale $x(t_0 + h)$:

$$x_1 := x_0 + h \, f(x_0, t_0) \simeq x(t_0+h)$$

$$x_2 := x_1 + h \, f(x_1, t_1)$$

con $t_1 := t_0 + h$

En general:

$$x_{n+1} := x_n + h \, f(x_n, t_n)$$



Vamos a resolver

$$\dot{x} = -x = f(x,t)$$

```julia
f(x,t) = -x
```

```julia
t0 = 0.0
x0 = 1.0

h = 0.1

x1 = x0 + h*f(x0, t0)
t1 = t0 + h
x2 = x1 + h*f(x1, t1)
```

```julia
t0 = 0.0
x0 = 1.0

h = 0.1

x_actual = x0
t_actual = t0


x_nueva = x_actual + h*f(x_actual, t_actual)
t_nueva = t_actual + h

println(t_nueva, "\t", x_nueva)

x_actual = x_nueva
t_actual = t_nueva
```

$$x_{n+1} := x_n + h \, f(x_n, t_n)$$


Envolver todo en una función:

```julia
function euler(f, t0, x0, t_final, h)

    x_actual = x0
    t_actual = t0

    ts = [t0:h:t_final]
    num_pasos = length(ts)

    xs = zeros(ts)

    for n in 1:num_pasos

        xs[n] = x_actual

        x_nueva = x_actual + h*f(x_actual, t_actual)
        t_nueva = t_actual + h

        #@show((x_nueva, t_nueva))

        x_actual = x_nueva
        t_actual = t_nueva
    end
    
    ts, xs
end
```

```julia
decaimiento(x,t) = -x
```

```julia
ts, xs = euler(decaimiento, 0, 1, 10, 0.5)
ts1, xs1 = euler(decaimiento, 0, 1, 10, 0.1)
```

```julia
plot(ts, xs)
plot(ts1, xs1)
```
