### Método de Euler con condición de paro

```julia
function euler(y0,t0,h,f,N,paro)
    y = Vector{Float64}[[y0]]
    t = Float64[t0]
    for i in 1:N
        y_nueva = y0 + h*f(t0, y0)
        t0 = t0 + h
        y0 = y_nueva
        push!(t,t0)
        push!(y,[y0])
        
        if paro(t0,y0) == true
            break
        end
    end
    t,hcat(y...)'
end
```

### Tiro parabólico con arrastre cuadrático

$$
\begin{array}{l}
\ddot{x} = -\lambda \|\vec{v}\| v_x \\
\ddot{y} = -\lambda \|\vec{v}\| v_y - g
\end{array}
$$

```julia
# Convención:
# Q[1] = x
# Q[2] = vx
# Q[3] = y
# Q[4] = vy
function tiro_parabolico(t,Q)
    λ = 0.1
    g = 9.8
    [Q[2], -λ*Q[2]*sqrt(Q[2]^2 + Q[4]^2), Q[4], -g-λ*Q[4]*sqrt(Q[2]^2 + Q[4]^2)]
end

function paro(t,Q)
    # El piso está en y = 0
    if Q[3] < 0
        true
    else
        false
    end
end
```

```julia
c_ini = [0, 30*cos(pi/3), 0, 30*sin(pi/3)]

t, M = euler(c_ini, 0, 0.001, tiro_parabolico, 15000, paro);
```

```julia
using PyPlot
```

```julia
plot(M[:,1], M[:,3]);
```

### Euler-Cromer

Separamos el sistema de ecuaciones en dos sistemas: uno para las coordenadas y otro para las velocidades.

$$
\begin{array}{l}
\dot{\vec{q}} = \vec{f}_{\vec{q}}(t,\vec{q}, \vec{v})\\
\dot{\vec{v}} = \vec{f}_{\vec{v}}(t,\vec{q}, \vec{v})
\end{array}
$$

El sistema de ecuaciones para las velocidades lo resolvemos con el método de Euler explícito y el sistema de ecuaciones para las coordenadas lo resolvemos con un método de Euler implícito.

\begin{array}{l}
v^k_{n+1} = v^k_n + hf^k_{v}(t_n,\vec{q}_n,\vec{v}_n)\\
q^k_{n+1} = q^k_n + hf^k_{q}(t_n,\vec{q}_n,\vec{v}_{n+1})
\end{array}

Por eso, el método de Euler-Cromer es **semi-implícito**.

```julia
function euler_cromer(q0,v0,t0,tf,h,f_q,f_v,paro)
    q = Vector{Float64}[]
    v = Vector{Float64}[]
    t = Float64[]
    for ti in t0:h:tf
        v_nueva = v0 + h*f_v(ti,q0,v0)
        q_nueva = q0 + h*f_q(ti,q0,v_nueva)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
        
        if paro(ti,q0,v0) == true
            break
        end
    end
    t,hcat(q...)',hcat(v...)'
end
```

```julia
# Convención:
# q[1] = x
# q[2] = y
# v[1] = vx
# v[2] = vy
function f_q(t,q,v)
    [v[1], v[2]]
end

function f_v(t,q,v)
    λ = 0.1
    g = 9.8
   [-λ*v[1]*sqrt(v[1]^2 + v[2]^2), -g-λ*v[2]*sqrt(v[1]^2 + v[2]^2)] 
end

function paro(t,q,v)
    # El piso están en y = 0
    if q[2] < 0
        true
    else
        false
    end
end
```

```julia
q_ini = [0, 0]
v_ini = [30, 70]

t, q, v = euler_cromer(q_ini, v_ini, 0, 10, 0.001, f_q, f_v,paro);
```

```julia
plot(q[:,1],q[:,2]);
```

### Péndulo simple

$$
\ddot{\theta} = -\frac{g}{\ell}\sin\theta
$$

```julia
function f_θ(t,θ,ω)
    ω
end

function f_ω(t,θ,ω)
    ℓ = 13.5
    -9.8/ℓ * sin(θ)
end

function paro(t,θ,ω)
    # El ángulo del péndulo simple va de -pi a pi
    if abs(θ) > pi
        true
    else
        false
    end
end

t,θ,ω = euler_cromer(pi-0.01,0,0,100,0.001,f_θ,f_ω,paro);
```

```julia
plot(θ,ω);
```

Debido a la no linealidad, la trayectoria en el espacio fase ya no es una elipse.

### Distintas condiciones iniciales

```julia
for θ_0 in -pi:pi/10:pi
    for ω_0 in -2:1:2
        t,θ,ω = euler_cromer(θ_0,ω_0,0,6,0.001,f_θ,f_ω,paro);
        plot(θ,ω)
        axis("equal")
    end
end
```

Las trayectorias con $|\omega| \gtrapprox 1.5$ corresponden a una frecuencia angular inicial tan grande que el péndulo da vueltas.

### El espacio fase del péndulo simple es un cilindro

```julia
# Dibujamos un cilindro transparente, alpha = 0.2

n = 50      # Discretización del eje z

zz = linspace(-5, 5, n)
radio = ones(n) #cilindro de radio uno   
theta = 2pi * linspace(0, 1, n)'

x = radio * cos(theta)
y = radio * sin(theta)
z = zz[:, ones(n)]   #  Tony's trick!   Who is Tony?  No idea.

cm = get_cmap("coolwarm")
surf = plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm,
linewidth=0, antialiased=true, alpha=0.2);

# Dibujamos tres trayectorias en el espacio fase
# r = 1, ϕ = θ, z = ω

t,θ,ω = euler_cromer(-pi,2,0,16,0.001,f_θ,f_ω,paro);
r = ones(length(θ))
curva_cilindrica(r, θ[:,1], ω[:,1])

t,θ,ω = euler_cromer(0,0.2,0,16,0.001,f_θ,f_ω,paro);
r = ones(length(θ))
curva_cilindrica(r, θ[:,1], ω[:,1])

t,θ,ω = euler_cromer(0,1.5,0,16,0.001,f_θ,f_ω,paro);
r = ones(length(θ))
curva_cilindrica(r, θ[:,1], ω[:,1]);
```
